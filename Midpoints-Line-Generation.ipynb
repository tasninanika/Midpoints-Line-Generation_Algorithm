```python
import matplotlib.pyplot as plt

def midpoint_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    d = 2 * dy - dx  
    y = y1
    points = [(x1, y1)]
    
    for x in range(x1, x2 + 1):
        points.append((x, y))
        if d < 0:
            d += 2 * dy  
        else:
            y += 1
            d += 2 * (dy - dx)  
    
    return points

def plot_line(points):
    x_vals, y_vals = zip(*points)
    plt.plot(x_vals, y_vals, marker='o', color='b', linestyle='-')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Midpoint Line Algorithm")
    plt.grid(True)
    plt.show()

# Example usage
x1, y1 = 2, 3
x2, y2 = 10, 7
points = midpoint_line(x1, y1, x2, y2)
plot_line(points)
```

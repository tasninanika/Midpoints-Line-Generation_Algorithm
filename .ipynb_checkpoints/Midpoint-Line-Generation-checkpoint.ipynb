```python
import matplotlib.pyplot as plt

def midpoint_line(x1, y1, x2, y2):
    x, y = x1, y1
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = 1 if x2 > x1 else -1  # Direction for x
    sy = 1 if y2 > y1 else -1  # Direction for y
    d = 2 * dy - dx  # Midpoint decision parameter
    points = [(x, y)]

    while x != x2:
        if d < 0:
            d += 2 * dy  # Choose E (east)
        else:
            y += sy  # Choose NE (northeast)
            d += 2 * (dy - dx)
        x += sx
        points.append((x, y))

    return points

# Example usage
x1, y1 = 2, 3
x2, y2 = 10, 7
points = midpoint_line(x1, y1, x2, y2)

# Extract x and y coordinates
x_vals, y_vals = zip(*points)

# Plot with connected lines directly
plt.plot(x_vals, y_vals, marker='o', color='green', linestyle='-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Midpoint Line Algorithm (Without plot_line function)")
plt.grid(True)
plt.show()
```
